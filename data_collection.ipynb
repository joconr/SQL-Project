# Getting Track Info from Official Spotify Indie Playlists (Spotify API)

### Track playlist_id changes according to the playlist I am gathering data from (Meaning I ran this three times according to three separate official indie playlists)

```python
# Import package to make web API requests
import requests
# Import json package to decode JSON
import json
import pandas as pd
from sqlalchemy import create_engine
```

```python
oauth_token = 'OAUTH-TOKEN'

headers = {'Authorization':f'Bearer {oauth_token}'}
```

```python
playlist_id = '37i9dQZF1DWYzpSJHStHHx'
```

```python
playlist_endpoint_url = f'https://api.spotify.com/v1/playlists/{playlist_id}'

playlist_api_request = requests.get(playlist_endpoint_url, headers=headers)
```

```python
playlist_api_response = json.loads(playlist_api_request.text)
```

```python
playlist_api_response['tracks']['items']
```

```python
tracks = playlist_api_response['tracks']['items']
```

```python
type(playlist_api_response['tracks']['items'])
```

```python
playlist_api_response['tracks']['items'][0]['track'].keys()
```

```python
track_dictionary = {
    'name':[],
    'artist':[],
    'album':[],
    'explicit':[],
    'popularity':[],
    'duration':[],
    'url':[]
}

for track in tracks:

    name = track['track']['name']
    track_dictionary['name'].append(name)
    print('Title:', name)

    artist = track['track']['artists'][0]['name']
    track_dictionary['artist'].append(artist)
    print('Primary Artist:', artist)

    album = track['track']['album']['name']
    track_dictionary['album'].append(album)
    print('Album:', album)

    explicit = track['track']['explicit']
    track_dictionary['explicit'].append(explicit)
    print('Explicit:', explicit)

    popularity = track['track']['popularity']
    track_dictionary['popularity'].append(popularity)
    print('Popularity Index:', popularity)

    duration = track['track']['duration_ms']
    track_dictionary['duration'].append(duration)
    print('Song Duration (in ms):', duration)

    url = track['track']['external_urls']['spotify']
    track_dictionary['url'].append(url)
    print('Spotify URL:', url)

    print('-'*75)
```

```python
track_dictionary
```

```python
playlist_track_df = pd.DataFrame(track_dictionary)
```
